## missing data

```python
total = data_train.isnull().sum().sort_values(ascending=False)
percent = (data_train.isnull().sum() / data_train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Totol', 'Percent'])
missing_data.head(20)
```

```python
data_test.shape
```

```python
data_train.drop((missing_data[missing_data.Totol > 1]).index, axis=1, inplace=True)
data_train.drop(data_train.loc[data_train.Electrical.isnull()].index, inplace=True)
data_train.isnull().sum().max()
```

## out liars

```python
from sklearn.preprocessing import StandardScaler
# np.newaxis分别是在行或列上增加维度，原来是（6，）的数组，在行上增加维度变成（1,6）的二维数组，在列上增加维度变为(6,1)的二维数组
saleprice_scaled = StandardScaler().fit_transform(data_train['SalePrice'][:,np.newaxis]) 
```

```python
saleprice_scaled.shape
```

```python
data_train.SalePrice.shape
```

```python
low_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][:10] # 价格从低到高并且按照index排序的前10个
high_range = saleprice_scaled[saleprice_scaled[:, 0].argsort()][-10:]
```

```python
print('outer range (low) of the distribution:')
print(low_range)
print('\nouter range (high) of the distribution:')
print(high_range)
```

## dummy variables

```python
y_train = data_train.SalePrice.values
data_train.drop(['SalePrice'], axis=1, inplace=True)
```

```python
train_objs_num = len(data_train)
dataset = pd.concat(objs=[data_train, data_test], axis=0, sort=True)
dataset_preprocessed = pd.get_dummies(dataset)
train_preprocessed = dataset_preprocessed[:train_objs_num]
test_preprocessed = dataset_preprocessed[train_objs_num:]
```

```python
x_train = train_preprocessed.drop(['Id'], axis=1).values
x_test = test_preprocessed.drop(['Id'], axis=1).values
```

## 随机森林

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=2000, oob_score=True, n_jobs=-1)
rf.fit(x_train, y_train)
```

```python
rf.score(x_train, y_train)
```

```python
cross_val_score(rf, x_train, y_train, cv=5).mean()
```

## XGBOOST

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
GBoost = GradientBoostingRegressor(n_estimators=3000)
GBoost.fit(x_train, y_train)
```

```python
GBoost.score(x_train, y_train)
```

```python
cross_val_score(GBoost, x_train, y_train, cv=5).mean()
```
