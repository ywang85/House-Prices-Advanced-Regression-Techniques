```python
y_train = train.SalePrice.values
train.drop(['SalePrice'], axis=1, inplace=True)

train_objs_num = len(train)
dataset = pd.concat(objs=[train, test], axis=0, sort=True)
dataset_preprocessed = pd.get_dummies(dataset)
train_preprocessed = dataset_preprocessed[:train_objs_num]
test_preprocessed = dataset_preprocessed[train_objs_num:]
```

```python
train_preprocessed.shape
```

```python
test_preprocessed.shape
```

```python
sum(train_preprocessed.isnull().sum())
```

```python
sum(test_preprocessed.isnull().sum())
```

```python
train_preprocessed = train_preprocessed.interpolate()
```

```python
test_preprocessed = test_preprocessed.interpolate()
```

## Build a linear model

```python
y = np.log(y_train)
X = train_preprocessed.drop(['Id'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=.33)
```

## Begin modelling

```python
from sklearn import linear_model
# 如果做了dummy，就不能用线性回归
lr = linear_model.LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
print ("R^2 is: \n", model.score(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print ('RMSE is: \n', mean_squared_error(y_test, predictions))
```

```python
actual_values = y_test
plt.scatter(predictions, actual_values)
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression')
plt.show()
```

## Try to improve the model

```python
for i in range (-2, 3):
    alpha = 10**i
    rm = linear_model.Ridge(alpha=alpha)
    ridge_model = rm.fit(X_train, y_train)
    preds_ridge = ridge_model.predict(X_test)

    plt.scatter(preds_ridge, actual_values, alpha=.75, color='b')
    plt.xlabel('Predicted Price')
    plt.ylabel('Actual Price')
    plt.title('Ridge Regularization with alpha = {}'.format(alpha))
    overlay = 'R^2 is: {}\nRMSE is: {}'.format(
                    ridge_model.score(X_test, y_test),
                    mean_squared_error(y_test, preds_ridge))
    plt.annotate(s=overlay,xy=(12.1,10.6),size='x-large')
    plt.show()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
predictions = rf.predict(X_test)
```

```python
print ('RMSE is: \n', mean_squared_error(y_test, predictions))
```

```python
actual_values = y_test
plt.scatter(predictions, actual_values)
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Randomforest')
plt.show()
```

```python
for i in range (1000, 6000, 1000):
    rf = RandomForestRegressor(n_estimators=i)
    rf.fit(X_train, y_train)
    preds = rf.predict(X_test)

    plt.scatter(preds, actual_values, alpha=.75, color='b')
    plt.xlabel('Predicted Price')
    plt.ylabel('Actual Price')
    plt.title('Random Forest with estimators = {}'.format(i))
    overlay = 'R^2 is: {}\nRMSE is: {}'.format(
                    rf.score(X_test, y_test),
                    mean_squared_error(y_test, preds))
    plt.annotate(s=overlay,xy=(12.1,10.6),size='x-large')
    plt.show()
```
